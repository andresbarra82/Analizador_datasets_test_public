```python
import pandas as pd
import numpy as np 

# Obtener los valores únicos de las variables categóricas

# Seleccionar las variables categóricas

valores_categoricos = df.select_dtypes(include=['object'])  # Seleccionar las variables categóricas

print(valores_categoricos)

valores_categoricos = pd.DataFrame(valores_categoricos) # Convertir las variables categóricas en un DataFrame

valores_categoricos_dataframe = valores_categoricos.nunique()

print(valores_categoricos_dataframe)



```

```python
import pandas as pd
import json

# Obtener los valores únicos de las variables categóricas

valores_unicos_por_categoria = {
    "valores_unicos": {
        columna: df[columna].unique().tolist()
        for columna in df.select_dtypes(include=['object']).columns
        if 'fecha' not in columna.lower()  # Filtra columnas que no contengan la palabra "fecha"
    }
}

with open('valores_unicos_contribuyentes', 'w') as f:      # Abrir un archivo JSON para guardar los valores únicos 'w' es para escribir with es para abrir y cerrar el archivo
    json.dump(valores_unicos_por_categoria, f, indent=4) # Guardar los valores únicos en un archivo JSON donde f es el archivo indent es la identación y dump es para guardar

valores_unicos_por_categoria
```
